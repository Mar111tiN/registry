### convert the N2 string to tube_df

```python
tank_string = r'(?:(?:Tank)?(?P<Tank>[?0-9TW]+)/)?'
turm_string = r'(?:(?:Turm)?(?P<Turm>[?0-9]+)/)?'
box_string = r'(?:(?:Box)?(?P<Box>[?0-9]+)/)?'
pos_string = r'Pos(?P<Pos>[-?0-9]+(?:,[-?0-9]+)*)'
cryo_pattern = re.compile(tank_string+turm_string+box_string+pos_string)

pos = r'Pos\.? ?(?P<Pos>[0-9]+(?:, ?[0-9]+)*); ?'
date = r'(?P<Date>[0-9][0-9][./-][0-9][0-9]?[./-][0-9][0-9])'
removed_pattern = re.compile(pos + date)


def convert_date(dates):
    # convert to str and strip 00:00:00
    
    date_str = dates.astype(str).str.replace('00:00:00', '').str.replace(' ', '')
    date_str = date_str.str.replace(r'.*([0-9][0-9])\.([0-9][0-9])\.([12]?[0-9]?[0-9][0-9]).*', r'\1.\2.\3')
    date_str = date_str.str.replace(r'.*([0-9][0-9])/([0-9][0-9])/([12]?[0-9]?[0-9][0-9]).*', r'\2.\1.\3')
    date_str = date_str.str.replace(r'.*([12][0-9][0-9][0-9])-([0-9][0-9])-([0-9][0-9]).*', r'\3.\2.\1')
    # add the 20 to 01 --> 2001
    date_str = date_str.str.replace(r'\.([0-9][0-9])$', r'.20\1')
    date = pd.to_datetime(date_str, format="%d.%m.%Y", exact=False)  # , errors='coerce'
    # date = pd.to_datetime(date_str, format="%d.%m.%Y", exact=False, errors='ignore')
    return date


def extract_cryo_loc(m, cryo_string, location={}):
    '''
    extracts cryo_info and returns location_dict and remaining string
    '''

    new_loc = {t:m.group(t) for t in ['Tank', 'Turm', 'Box', 'Pos'] if m.group(t)}
    location.update(new_loc)
    rest_string = cryo_string[m.end():]
    return rest_string, location


def make_cryo_tube_df(loc_dict):   
    loc_dict['Pos'] = loc_dict['Pos'].split(',')
    df = pd.DataFrame.from_dict(loc_dict)
    df['Removed'] = (df['Pos'] == "-").astype(int)
    
    # adjust removed and unknown samples
    df['Pos'] = df['Pos'].replace(['?', '-'], -50).astype(int)
    while len(df.loc[df['Pos'] < 0, 'Pos']):
        df.loc[df['Pos'] < 0, 'Pos'] = df.shift(-1)['Pos'] - 1
    df['Pos'] = df['Pos'].fillna(-50).astype(int)
    while len(df.loc[df['Pos'] < 0, 'Pos']):
        df.loc[df['Pos'] < 0, 'Pos'] = df.shift(1)['Pos'] + 1
    df['Pos'] = df['Pos'].astype(int)
        
    return df


def cryo2df(row, col='N2'):
    '''
    consume 
    '''
    cryo_string = row[col] # 'N2'
    loc_dict={}
    df = pd.DataFrame()
    while (m := cryo_pattern.match(cryo_string)):
        cryo_string, loc_dict = extract_cryo_loc(m, cryo_string, loc_dict)
        df = pd.concat([df, make_cryo_tube_df(loc_dict)])
    return df.reset_index(drop=True)

def removed2df(row, col='N2removed'):
    '''
    make a removed df from N2removed
    '''
    
    if (m:= removed_pattern.match(row[col])):
        # make a tube list from positions 
        pos = m.group('Pos').replace(' ','').split(',')
        df = pd.DataFrame.from_dict({'Pos':pos, 'removeDate': m.group('Date')})
    else:
        df = pd.DataFrame(columns=['Pos', 'removeDate'])
    df['Pos'] = df['Pos'].astype(int)
    return df
```

### convert the count string to df

```python
type_pattern = r'(?P<Type>[A-Za-z\-]+)'
cell_count_pattern = r'(?P<Count>[0-9]+(?:,[0-9]+)?[Ee][0-9])?'
tube_count_pattern = r'\((?P<Tubes>[0-9]+)\)'
count_pattern = re.compile(type_pattern + cell_count_pattern + tube_count_pattern)


def extract_count(m, count_string):
    '''
    extracts cryo_info and returns location_dict and remaining string
    '''
    # print(count_string)
    count_dict = {t:m.group(t) for t in ['Type', 'Count', 'Tubes']}
    rest_string = count_string[m.end():]
    return rest_string, count_dict


def make_count_tube_df(count_dict):
    count_dict['Type'] = [count_dict['Type'] for tube in range(int(count_dict['Tubes']))]
    df = pd.DataFrame.from_dict(count_dict)
    return df


def count2df(row, col='Count'):
    count_string = row[col] # 'Count'
    
    df = pd.DataFrame()
    while (m := count_pattern.match(count_string)):
        count_string, count_dict = extract_count(m, count_string)
        df = pd.concat([df, make_count_tube_df(count_dict)])
    return df.reset_index(drop=True).drop(columns='Tubes')
```

### merge the dfs

```python
def get_tube_df(row, count_col='Count', cryo_col='N2', removed_col='N2removed'):
    '''
    creates tube_df from all info in a row of sample_df
    '''
    
    # create the count_dfs
    cryo_df = cryo2df(row, col=cryo_col)
    removed_df = removed2df(row, col=removed_col)
    count_df = count2df(row, col=count_col)
    # merge the tubes to complete the tube positions
    merged1 = cryo_df.merge(removed_df, on='Pos', how="outer")
    # merge the type info
    merged2 = merged1.merge(count_df, left_index=True, right_index=True, how="outer")
    
    # edit the data
    # sort                                        
    tube_df = merged2.sort_values(['Tank', 'Turm', 'Box', 'Pos'])
    # fill in missing info
    tube_df = tube_df.fillna(method='bfill').fillna(method='ffill')
    # convert to datetime
    tube_df['removeDate'] = convert_date(tube_df['removeDate'])
    tube_df['Removed'] = tube_df['Removed'].astype(int)
    tube_df['SampleID'] = row.name
    return tube_df
```

```python
# set column constants
patient_cols = ['PatientID', 'LastName','FirstName', 'DOB', 'SAP', 'Clinix']
data_cols = ['Timepoint', 'Date', 'Type', 'Count', 'N2', 'N2removed', 'DNA', 'Note']  # columns that are repeated horizontally and need to be reduced

project = 'Project'
sample_cols = ['Sample_ID', timepoint, 'Date', 'Organ'] + data_cols
cryo_cols = ['Cryo_ID', 'Tank', 'Tower', 'Box', 'Position']
cryo_info = ['Organ', 'Population', 'Count']
# regex patterns
count_pattern = re.compile(r'[^MmGgPpKk0-9]*(?:(?P<Organ>[A-Za-z])/)?(?P<Population>[A-Za-z]+)?[^0-9]?(?P<Cells>[,0-9xEe;]+)?[^\(]*\((?P<Count>[0-9]+)\)[^MmGgPpKk0-9]*(?P<rest>[A-Za-z]+.*)?')
tank_string = '[^Tt]*(?P<tank>Tank ?[^/]*)[ /,]+'
turm_string = 'Turm ?(?P<turm>\d+)[ /,]+'
box_string = 'Box ?(?P<box>\d+)[ /,]+[^-0-9]*'
pos_string = '(?P<pos>[-0-9, ]+)[ /,;]*(?P<rest>[TtBb].+)?'

tank_pattern = re.compile(tank_string + turm_string + box_string + pos_string)
turm_pattern = re.compile(turm_string + box_string + pos_string)
box_pattern = re.compile(box_string + pos_string)


def retrieve_tubes(row):
    '''
    expands every row of sample_df to a dataframe of cryo_tubes
    '''
    # get strings
    count_string = row['Count'] # 'Count'
    cryo_string = row['N2'] # 'N2'
    # dna_string = row[data_cols[3]] # 'DNA'
    
    if cryo_string == 'nan' or cryo_string == '/':
        return pd.DataFrame().append(row)
    ########
    print(row.name, row['Patient_ID'], count_string, cryo_string)
    ########
    
    def make_pos_df(tank,turm,box,pos_string):
        '''
        returns a position dataframe parsed from cryostring
        '''
        # initialize dataframe
        df = pd.DataFrame()
        # make a row for each position
        
        positions = [pos for pos in pos_string.replace(' ','').replace('-',',-,').split(',') if pos]
        for pos in positions:
            row[cryo_cols[-1]] = pos
            df = df.append(row)
        df[cryo_cols[-4]] = tank
        df[cryo_cols[-3]] = turm
        df[cryo_cols[-2]] = box
        
        # for i in list(df.index)[-2::-1]:
          #   if df.iloc[i,7] == '-' and not df.iloc[i+1,7] == '-':
            #     df.iloc[i,7] = int(df.iloc[i+1,7]) - 1
              #   df.iloc[i,5] = df.iloc[i,5] + ' <<Position inferred>>'
        return df
        
    ###################################################################################################
    #######################################  PARSING N2  ##############################################
    m = tank_pattern.match(cryo_string)
    tank = m.group('tank')
    turm = m.group('turm')
    box = m.group('box')
    
    # make the pos_df
    print(m.group('pos'))
    pos_df = make_pos_df(tank, turm, box, m.group('pos'))
    rest_string = m.group('rest')
  
    # handle rest_string (has to contain at least numbers)
    if rest_string and bool(re.search(r'\d', rest_string)):
        ##########
        # print(rest_string)
        ##########
        if 'Turm' in rest_string:
            m = turm_pattern.match(rest_string)     
            turm = m.group('turm')
            box = m.group('box')
            pos_df = pd.concat([pos_df, make_pos_df(tank, turm, box, m.group('pos'))], sort=True)
        elif 'Box' in rest_string:
            m = box_pattern.match(rest_string)
            box = m.group('box')
            pos_df = pd.concat([pos_df, make_pos_df(tank, turm, box, m.group('pos'))], sort=True)
       
    # remove superfluous columns to prevent doubling data
    pos_df = pos_df.reset_index().drop(sample_cols[1:4], axis=1)
    #######################################  PARSING N2  ##############################################
    ###################################################################################################
    
    ###################################################################################################
    #######################################  PARSING COUNT  ###########################################
    tube_df = pd.DataFrame()
    
    # account for 'nicht gezählt'
    if count_string == 'I nicht gezählt' or count_string == '' or count_string == 'nan' or count_string == '?':
        for _ in range(len(pos_df.index)):
            cryo_row = pd.Series({
                cryo_info[0]: None,
                cryo_info[1]: None,
                cryo_info[2]: None
            })
            tube_df = tube_df.append(cryo_row, ignore_index=True)
        count_string = ''
    while count_string: 
        m = count_pattern.match(count_string)
        if m:
            count_string = m.group('rest')
            organ = m.group('Organ')
            population = m.group('Population')
            cell_count = m.group('Cells')
            cryo_count = int(m.group('Count'))
        
            # infer stuff   
            if cell_count and not population:
                population = 'MN'
            if cell_count:
                population = 'MN'
            # infer organs if not set in sample
            if not row['Organ'].upper() in ['PB', 'BM','KM']:
                if 'GR' in population.upper() or 'PL' in population.upper():
                    organ = 'PB'
                if population == 'MN' and not organ:
                    organ = 'PB'
                    
            # return row for matches
            cryo_row = pd.Series({
                cryo_info[0]: organ,
                cryo_info[1]: population,
                cryo_info[2]: cell_count
            })
            for pos in range(cryo_count):
                tube_df = tube_df.append(cryo_row, ignore_index=True)
                
            ###############################################
            # print(row.name, organ, population, cell_count)
            ###############################################
        else:
            raise ValueError('No match with Count pattern')
    #######################################  PARSING COUNT  ###########################################
    ###################################################################################################
    

    # check for same size of dfs --> pass message to Note
    if len(pos_df.index) != len(tube_df.index):
        pos_df['Note'] = pos_df['Note'] + ' <<Check tube count!>>'
    
    # merge pos_df and tube_df
    merge_df = pos_df.merge(tube_df, left_index = True, right_index=True, how='inner')
    
    if (merge_df['Position'] == '-').any():
        # infer positions from successor position
        for i in list(merge_df.index)[-2::-1]:
            if merge_df.iloc[i][7] == '-' and not merge_df.iloc[i+1,7] == '-':
                merge_df.iloc[i,7] = int(merge_df.iloc[i+1,7]) - 1
                merge_df.iloc[i,5] = merge_df.iloc[i,5] + ' <<Position inferred>>'
    
        # infer positions from predecessor position
        for i in list(merge_df.index)[:-1]:
            if merge_df.iloc[i][7] != '-' and merge_df.iloc[i+1,7] == '-':
                merge_df.iloc[i+1,7] = int(merge_df.iloc[i,7]) + 1
                merge_df.iloc[i+1,5] = merge_df.iloc[i+1,5] + ' <<Position inferred>>'

    # format Sample_ID and Cryo_ID
    ##########
    ##########
    merge_df[sample_cols[0]] = merge_df[sample_cols[0]].astype('int')
    # create cryo_ID
    merge_df[cryo_cols[0]] = merge_df[sample_cols[0]].astype('str') + '-' + merge_df.index.astype('str') 

    
    # format integer columns
    merge_df.loc[merge_df['Position'] == "-"]['Position'] = -1
    
    for col in cryo_cols[2:-1]:
        merge_df[col] = merge_df[col].astype('int')  
    
    return merge_df  
```
